```python
import cv2
```

```python
#Loading Mask RCNN
net = cv2.dnn.readNetFromTensorflow("Pictures/Python/dnn/frozen_inference_graph_coco.pb",
                                    "Pictures/Python/dnn/mask_rcnn_inception_v2_coco_2018_01_28.pbtxt")
```

```python
#Load an image
img = cv2.imread("Pictures\Saved Pictures\Good1.jpg")
height, width, _ = img.shape

cv2.imshow("pic",img)
cv2.waitKey(0)
```

```python
#Detect Objects
blob = cv2.dnn.blobFromImage(img, swapRB=True)
net.setInput(blob)
boxes, masks = net.forward(["detection_out_final", "detection_masks"])
detection_count = boxes.shape[2]

```

```python
#Extract Information

for i in range(detection_count):
    box = boxes[0,0,i]
    print(box)
    x = int(box[3] * width)
    y = int(box[4] *height)
    x2 = int(box[5] * width)
    y2 = int(box[6] *height)

    cv2.rectangle(img, (x,y), (x2,y2), (255, 0, 0))


cv2.imshow("pic",img)
cv2.waitKey(0)

```
